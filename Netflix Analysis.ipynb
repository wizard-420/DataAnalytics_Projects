## Analyze the data and generate insights that could help Netflix in deciding which type of shows/movies to produce and how they can grow the business in different countries

## Handling Nulls

```python
sample1['director'].fillna('Unknown Director',inplace = True)
sample1['cast'].fillna('Unknown Cast',inplace = True)
sample1['country'].fillna('Unknown Country',inplace = True)
#sample1['date_added'] = sample1['date_added'].fillna(0,inplace = True)
sample1.dropna(subset=['rating'],inplace=True)
sample1['duration'].fillna('0',inplace = True)
sample1.dropna(subset=['date_added'],inplace=True)
sample1['date_added'] = pd.to_datetime(sample1['date_added'])
sample1.info()
```

```python
#Best countries for movie hosting.
movies['country'] = movies['country'].str.replace(" ","")
movies['country'] = movies['country'].str.replace("UnitedStates","U.S.A")
movies['country'] = movies['country'].str.replace("UnitedKingdom","U.K.")
country_count = pd.DataFrame(movies['country'].str.split(",").explode())
count = country_count[country_count['country']!='UnknownCountry'].groupby('country')['country'].count().sort_values(ascending = False)
top20 = count[0:20]
top20.plot(kind = 'bar')
```

## Popularity of types(Tv shows,movies) in diffrent countries

```python
sample1['country'] = sample1['country'].str.replace(" ","")
sample1['country'] = sample1['country'].str.replace("UnitedStates","U.S.A")
sample1['country'] = sample1['country'].str.replace("UnitedKingdom","U.K.")
sample1['country'] = sample1['country'].str.split(",")
df2 = sample1.explode('country')
df2['country'].value_counts()
```

```python
top10 = df2['country'].value_counts().sort_values(ascending = False)[0:10]
plt.figure(figsize=(10,10))
sns.countplot(data = df2[(df2['country'].isin(top10.index))&(df2['country']!='UnknownCountry')],x = 'country',hue='type')
plt.show()
```

Among top 10 countries of content producers in netflix only Japan and South Korea prefer TV shows more then movies.

In India 'bollywood' is big buisness hence we can see the content produced is mostly movies as compared to Tv shows.
Same can be concluded for USA as majority of the content creation buisness comes from 'Hollywood'

As other studies suggest 'K dramas' are more popular in South Korea and 'Anime' in Japan. Both are in category of TV shows.

## Top Movie and TV shows director

```python
sample1['director'] = sample1['director'].str.split(",")
df3 = sample1.explode('director')
movie_directors = df3[(df3['type']=='Movie')&(df3['director']!='Unknown Director')].groupby('director')['title'].count().sort_values(ascending = False)[0:15]
movie_directors.plot(kind = 'bar')
```

top TV shows directors
